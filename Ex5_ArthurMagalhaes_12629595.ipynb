## Exercício 4

Examine a correlação entre a cor (u-r) com a idade (logt) e com a metalicidade (logZ) para a amostra em novais.txt. O que você conclui? Discuta se e como esse resultado depende do método usado para medir a correlação.

## Discussão

```r
tab = read.table('./novais.txt', head = TRUE)
head(tab)


plot(tab$umr,tab$logt, pch=19, ylab= 'idade',xlab= 'cor')
plot(tab$umr,tab$logZ, pch=19, ylab= 'metalicidade', xlab='cor')
```

## Teste de correlação (u-r) x (logt)

```r
cor.test(tab$umr,tab$logt,method = 'pearson')
cor.test(tab$umr,tab$logt,method = 'spearman')
cor.test(tab$umr,tab$logt,method = 'kendall')
```

## Teste de correlação (u-r) x (logZ)

cor.test(tab$umr,tab$logZ,method = 'pearson')
cor.test(tab$umr,tab$logZ,method = 'spearman')
cor.test(tab$umr,tab$logZ,method = 'kendall')

## Discussão

Aplicando os diferentes testes de correlação para as medidas, podemos concluir que existe uma forte correlação entre a cor e a idade, e também um a correlação, embora menos forte, entre a cor e a metalicidade

O coeficiente de Kendall, ou coeficiente $\tau$ de Kendall é uma grandeza utiliada na estatística para medir associação ordinária entre 2 medidas mensuráveis.

Ele fornece o grau de similaridade entre dois conjuntos, dado o mesmo objetos. Básicamente ele mostra quantas alterações devem ser fetias para um conjunto fique igual ao outro.
